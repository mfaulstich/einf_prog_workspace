# Reihenuntersuchung – 1000 Partialsummen

In diesem Notebook werden vier klassische Reihen untersucht:

1. **Harmonische Reihe:**  
$$H_n = \sum_{k=1}^{n} \frac{1}{k}$$

2. **Alternierende harmonische Reihe:**  
$$A_n = \sum_{k=1}^{n} (-1)^{k+1}\frac{1}{k}$$

3. **Leibnizsche Reihe (für $\pi/4$):**  
$$L_n = \sum_{k=0}^{n-1} (-1)^k \frac{1}{2k+1}$$

4. **Geometrische Reihe (mit $r = 1/2$):**  
$$G_n = \sum_{k=0}^{n-1} \left(\frac{1}{2}\right)^k$$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

# Harmonische Reihe
H = np.cumsum(1 / np.arange(1, N+1))

# Alternierende harmonische Reihe
A = np.cumsum(((-1)**(np.arange(1, N+1)+1)) / np.arange(1, N+1))

# Leibnizsche Reihe (pi/4)
L = np.cumsum(((-1)**np.arange(N)) / (2*np.arange(N) + 1))

# Geometrische Reihe (r = 1/2)
G = np.cumsum((1/2)**np.arange(N))

H[-1], A[-1], L[-1], G[-1]
```

Die letzten Werte (für $n = 1000$) sind:

- Harmonische Reihe: Divergent, wächst langsam  
- Alternierende harmonische Reihe: $\approx \ln(2) = 0.6931$  
- Leibnizsche Reihe: $\approx \frac{\pi}{4} = 0.7854$  
- Geometrische Reihe: $\to 2$

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0,0].plot(H, color='C0')
axs[0,0].set_title('Harmonische Reihe')
axs[0,0].set_xlabel('n')
axs[0,0].set_ylabel('H_n')

axs[0,1].plot(A, color='C1')
axs[0,1].axhline(np.log(2), color='gray', linestyle='--', label='ln(2)')
axs[0,1].set_title('Alternierende harmonische Reihe')
axs[0,1].legend()

axs[1,0].plot(L, color='C2')
axs[1,0].axhline(np.pi/4, color='gray', linestyle='--', label='π/4')
axs[1,0].set_title('Leibnizsche Reihe')
axs[1,0].legend()

axs[1,1].plot(G, color='C3')
axs[1,1].axhline(2, color='gray', linestyle='--', label='Grenzwert 2')
axs[1,1].set_title('Geometrische Reihe (r = 1/2)')
axs[1,1].legend()

plt.tight_layout()
plt.show()
```
