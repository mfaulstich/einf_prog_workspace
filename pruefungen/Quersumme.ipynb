```python
#
# Quersumme
# 

#####################################################
# Lösung Studierende
#####################################################

# BEGIN Vorbelegung des Antwortfeldes
def cross_sum(n):
    # Ihr Code hier
# END Vorbelegung des Antwortfeldes
    if (not type(n) == str):
        raise ValueError(f"Kein String übergeben, sondern {type(str)}")
    cross_sum = 0
    for c in n:
        try:
            cross_sum += int(c)
        except ValueError:
            raise ValueError(f"String enthält Zeichen {c}, das keine Ziffer ist.")
# BEGIN Vorbelegung des Antwortfeldes
    return cross_sum
# END Vorbelegung des Antwortfeldes


```
