```python
#
# Feld mischen
# 

#####################################################
# Lösung Studierende
#####################################################

from  random import Random

# BEGIN Vorbelegung des Antwortfeldes
def shuffle(field, count):
    rd = Random()
    for i in range(count):
        done = False
        while not done:
            idx1 = rd.randint(0, len(field)-1)
            idx2 = rd.randint(0, len(field)-1)
            if (idx1 == idx2):
                continue
            done = True
            tmp=field[idx1]
            field[idx1] = field[idx2]
            field[idx2] = tmp
            
        


```

```python
import math

class MatchError(ValueError):
    pass

def test_shuffle():
    # Feldlängen
    for fieldlen in range(10, 1000, 10):
        max=0;
        changes = int(fieldlen/2)
        f_org = []
        f_shuf = []
        for i in range(int(fieldlen)):
            f_org.append(i)
            f_shuf.append(i)
        shuffle(f_shuf, changes)
        no_match = 0
        for i in range(fieldlen):
            if (f_org[i] != f_shuf[i]):
                no_match += 1;
        max = max if max > no_match else no_match
    
        if no_match <= 0:
            raise MatchError("Keine Vertauschung")
        if int(no_match/2) > changes:
            raise MatchError(f"Zu viele Vertauschungen. {int(no_match/2)} bei Vorgabe {changes}")
        # Die Hälfte der Vertauschungen muss mindestens vorhanden sein.
        # Berücksichtigt Zurücktauschungen.
        if no_match < changes: 
            raise MatchError(f"Zu wenige Vertauschungen. {int(no_match/2)} bei Vorgabe {changes}")
        # Listen immer noch mit den gleichen Einträgen?
        for i in f_org:
            if not i in f_shuf:
                raise MatchError(f"Element {i} fehlt nach Vertauschungen.")
        
    print("OK", end="")
        
test_shuffle()
```
