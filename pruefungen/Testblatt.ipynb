```python
def berechne_untermatrix(matrix, spalte):
    """Entfernt die erste Zeile und die angegebene Spalte für die Laplacesche Entwicklung."""
    return [reihe[:spalte] + reihe[spalte+1:] for reihe in matrix[1:]]

def berechne_determinante(matrix):
    """Berechnet die Determinante einer quadratischen Matrix rekursiv."""
    if len(matrix) == 1:
        return matrix[0][0]
    if len(matrix) == 2:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
    
    det = 0
    for spalte in range(len(matrix)):
        vorzeichen = (-1) ** spalte
        unter_det = berechne_determinante(berechne_untermatrix(matrix, spalte))
        det += vorzeichen * matrix[0][spalte] * unter_det
    return det

def n_dimensionales_kreuzprodukt(vektoren):
    """
    Berechnet das Kreuzprodukt von (n-1) Vektoren im n-dimensionalen Raum.
    Eingabe: Liste von Listen/Tupeln (die Vektoren).
    """
    n = len(vektoren) + 1  # Die Dimension n
    ergebnis = []
    
    for i in range(n):
        # Erstelle eine Untermatrix durch Entfernen der "virtuellen" Einheitsvektor-Zeile
        # und der i-ten Spalte
        untermatrix = []
        for v in vektoren:
            reduzierter_vektor = list(v[:i]) + list(v[i+1:])
            untermatrix.append(reduzierter_vektor)
            
        # Der Wert an Stelle i ist die Determinante der Untermatrix (mit Vorzeichenwechsel)
        vorzeichen = (-1) ** i
        ergebnis.append(vorzeichen * berechne_determinante(untermatrix))
        
    return ergebnis

# Beispiel 1: Klassisches 3D Kreuzprodukt (2 Vektoren benötigt)
v1 = [1, 2, 3]
v2 = [4, 5, 6]
print(f"3D Kreuzprodukt: {n_dimensionales_kreuzprodukt([v1, v2])}")
# Erwartet: [-3, 6, -3]

# Beispiel 2: 4D Kreuzprodukt (3 Vektoren benötigt)
v3 = [1, 0, 0, 0]
v4 = [0, 1, 0, 0]
v5 = [0, 0, 1, 0]
print(f"4D Kreuzprodukt: {n_dimensionales_kreuzprodukt([v3, v4, v5])}")
# Erwartet: [0, 0, 0, 1]
```
