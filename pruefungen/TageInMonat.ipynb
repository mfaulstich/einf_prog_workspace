```python
def schaltjahr(jahr):
    schaltjahr = jahr % 4 == 0
    if (jahr % 100 == 0):
        return jahr % 400 == 0
    else:
        return schaltjahr
    
def tage_in_monat(monat, jahr):
    match monat:
        case 1 | 3 | 5 | 7 | 8 |10 | 12:
            return 31
        case 2:
            return 29 if schaltjahr(jahr) else 28
        case _:
            return 30
        
    
```
