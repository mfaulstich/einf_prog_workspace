#### 3.1. Feld mit Zufallszahlen belegen – Kennwerte ermitteln

Schreiben Sie ein Python-Methode, in welchem ein Feld mit Zufallszahlen zwischen
0 und 50 (jeweils einschließlich) belegt und zurück gegeben wird. Die Anzahl der zu belegenden Feldelemente wird als Parameter übergeben. Die Funktion `randint(minimum, maximum` aus dem Package `import random`, liefert Zufallszahlen im Bereich von minimum bis maximum, einschließlich. 

Geben Sie das so belegte Feld mit 10 Werten pro Zeile aus.<br>
* `print("...")` fügt an den ausgegebenen String automatisch einen Zeilenvorschub an, es sei denn, dem String wird ein Parameter<br>
`print("..." end = ?` angehangen. Der Wert, dem end an Stelle von`?` zugewiesen wird, kann ein String oder ein einzelnes Zeichen sein. Dieser String / dieses Zeichen wird als Abschluss der Zeile verwendet.
* Durch eine **bedingte Zuweisung** `value = x if Bedingung else y` kann entweder ein Zeilenvorschub-Zeichen `'\n'` für das Ende einer Zeile oder ein `", "` als Trennzeichen zwischen zwei ausgegebenen Zufallszahlen in einer Zeile verwendet werden. Die Bedingung wiederum kann so aufgebaut werden:<br>
  * Der Ausdruck `i % count` wird 0, wenn i ganzzahlig durch count teilbar ist.<br>
  * count ist hier die Anzahl der Werte pro Zeile.<br>
  * DANN müssen Sie sich nur noch etwas einfallen lassen, damit nicht schon nach dem ersten Zeichen Wert ein Zeilenumbruch erfolgt.

* Bilden Sie das arithmetische Mittel aller Feldelemente in einer weitern Methode und geben Sie diesen Wert aus.
* Ermitteln Sie den kleinsten Wert (Minimum) und den größten Wert (Maximum) der Feldelemente mit weiteren Methoden und geben Sie diese beiden Werte aus.

Die Methoden zur Ausgabe, und zur Berechnung von Mittelwert, Minimum und Maximum übernehmen zur Berechnung das Feld als Parameter.

Vorgehensweise: die Variablen, die die beiden Extremwerte aufnehmen sollen, werden mit
dem ersten Wert der zu untersuchenden Menge initialisiert. Dann erfolgt ein fortlaufender Vergleich mit allen anderen Werten der Menge. Wird ein größerer beziehungsweise kleinerer Wert gefunden, dann wird dieser als neuer Extremwert genommen.
Führen Sie die Berechnung in einem ersten Versuch zunächst mit drei oder vier Zahlen durch und kontrollieren Sie die Ergebnisse !



````python
import random


def rand_field(count):
    ### BEGIN SOLUTION
    a = []
    for i in range(0, count):
        a.append(random.randint(0, 50))
    ### END SOLUTION
    return a


def print_field(field):
    ### BEGIN SOLUTION
    idx = 0
    for value in field:
        print(f"{value:2}", end=", " if (idx + 1) % 4 != 0 else "\n")
        idx += 1


### END SOLUTION


def min_value(field):
    ### BEGIN SOLUTION
    min = field[0]
    for value in field:
        if value < min:
            min = value
    ### END SOLUTION
    return min


def max_value(field):
    ### BEGIN SOLUTION
    max = field[0]
    for value in field:
        if value > max:
            max = value
    ### END SOLUTION
    return max


def mean_value(field):
    ### BEGIN SOLUTION
    count = 0
    sum = 0
    for value in field:
        sum += value
        count += 1
    mean = sum / float(count)
    ### END SOLUTION
    return mean


field = rand_field(100)
print_field(field)
print(f"Mittelwert = {mean_value(field)}")
print(f"Minimum    = {min_value(field)}")
print(f"Maximum    = {max_value(field):2}")
````

````python
### BEGIN HIDDEN TESTS
assert len(rand_field(100)) == 100
assert min_value([9, 1000, -5, 20000, 8]) == -5
assert max_value([9, 1000, -5, 20000, 8]) == 20000
assert mean_value([-14, 3, 6, 9, 201]) == 41
### END HIDDEN TESTS
````

#### 3.2. Kennwerte einer Zahlenreihe aus einer Datei
Schreiben Sie auf der Basis des Programmes aus dem vorhergehenden Abschnitt 3.1 ein Programm,
welches eine unbekannte Anzahl von Gleitkommazahlen aus der Datei daten-3-2.txt einliest (siehe
Skript, Kapitel 1).
Die Anzahl der eingelesenen Werte (ein Wert pro Zeile) wird in der Einleseschleife ermittelt und aus-
gegeben. Geben Sie am Programmende den kleinsten Wert, den größten Wert und das arithmetische
Mitte aller Werte aus.

Sie benötigen dazu:
|Anweisung | Beschreibung |
|---|---|
| `with open(filename, "r") as file:` | Block, in dem die Date mit dem Namen in `filename` gelesen wird. Mit `file`wird auf die Datei zugegriffen.
`content = file.read()`  | In dem Block der geöffneten Datei `file`kann so der Inhalt der Datei in die Variable `content` gelesen werden. |
| `items = content.split()` | gibt eine Liste aller Einträge (Wörter, Zahlen) in dem gelesenen Inhalt `content` zurück, die durch Whitespace-Zeichen (Leerzeichen, Tabulator, Zeilenvorschub) getrennt sind.|
| `float(item)` | Konvertiert einen einzelnen Eintrag aus der Liste der Zahlen in eine Fließkommazahl. |


Führen Sie den Code zum Behandeln möglicher Fehler zwischen `try:` und `except FileNotFoundError:` in folgendem Block aus.

````python
def read_field(filename):
    ### BEGIN SOLUTION
    a = []
    with open(filename, "r") as file:
        content = file.read()
        count = 1
        for item in content.split():
            try:
                number = float(item)
                a.append(number)
                count += 1
            except ValueError:
                raise ValueError(
                    f"Ungültige Zahl {item} an Position {count} in Datei  {filename} gefunden."
                )
    ### END SOLUTION
    return a


filename = "daten-3-2.txt"
try:
### BEGIN SOLUTION
    field = read_field(filename)
    print(f"Anzahl Werte = {len(field)}")
    print(f"Mittelwert   = {mean_value(field)}")
    print(f"Minimum      = {min_value(field)}")
    print(f"Maximum      = {max_value(field):2}")
### END SOLUTION
except FileNotFoundError:
    print("Fehler!")
    print(f"Die Datei '{filename}' wurde nicht gefunden.")
except PermissionError:
    print("Fehler!")
    print(f"Auf die Datei '{filename}' kann nicht zugegriffen werden.")
except ValueError as ve:
    print("Fehler beim Konvertieren einer Zahl!")
    print(str(ve))
except Exception as e:
    print("Unwerwarteter Fehler!")
    print(str(e))
````

#### 3.3. Korrespondierende Felder
 
Schreiben Sie ein Python-Skript, in welchem die double-Felder x, y1 und y2 mit jeweils 1000 Elementen
angelegt werden. Die Funktionen y1 (x) = sin(x) und y2(x) = cos(x) sind im Bereich von x = 0 . . . 6.4
auszuwerten (include-Datei cmath ). Dazu wird das Intervall [0, 6.4] in 100 Teilintervalle der Länge
∆x = 6.4/100 eingeteilt und die beiden Funktionen an den 101 Teilungspunkten ausgewertet (siehe
Aufgabe 2.3). Die x-Werte werden fortlaufen im Feld x, die Funktionswerte entsprechend in den
Feldern y1 und y2 abgelegt.
Nachdem die Felder belegt sind, sollen die Werte zeilenweise in der Form

```
+0.000000  +0.000000  +1.000000
+0.064000  +0.063956  +0.997953
+0.128000  +0.127651  +0.991819
...
```
in die Datei `kfeld.dat` Datei ausgegeben werden.<br>
Überprüfen Sie, ob alle Werte vorhanden und richtig berechnet sind !

````python
import math

# Parameter
x_start = 0.0
x_end = 6.4
count_interval = 100
filename = "kfeld.dat"

try:
    ### BEGIN SOLUTION
    # Schrittweite berechnen
    dx = (x_end - x_start) / count_interval

    # Liste (Feld) für alle Werte: [ [x, y1, y2], ... ]
    values = []

    # Schleife über alle Stützstellen
    for i in range(count_interval + 1):  # +1, um auch x = 6.4 einzuschließen
        x = x_start + i * dx
        y1 = math.sin(x)
        y2 = math.cos(x)
        values.append([x, y1, y2])

    # Ergebnisse in Datei schreiben
    with open(filename, "w") as file:

        # Werte zeilenweise ausgeben
        for x, y1, y2 in values:
            file.write(f"{x:10.5f} {y1:15.8f} {y2:15.8f}\n")

    print(f"{len(values)} Wertepaare wurden in '{filename}' gespeichert.")
### END SOLUTION
except FileNotFoundError:
    print("Fehler!")
    print(f"Die Datei '{filename}' wurde nicht gefunden.")
except PermissionError:
    print("Fehler!")
    print(f"Auf die Datei '{filename}' kann nicht zugegriffen werden.")
except Exception as e:
    print("Unwerwarteter Fehler!")
    print(str(e))
````

Zeichen Sie mit `pyplotmatplotlib.pyplot` den Graphen<br>

in den Grenzen von $x=0$ bis $x=6.4$<br>
mit 1000 **Intervallen** (das ist <u>nicht</u> die Anzahl der erforderlichen x-Werte!). 

* Erstellen Sie zunächst eine Liste mit den erforderlichen x-Werten.
* Berechnen Sie die zugehörigen Werte für $\sin(2x)$ in einer weiteren Liste (`y1)`. Zur Verwendung der Methode `math.sin(...)` muss das erforderlichen Package importiert werden: `import math`.
* Berechnen Sie die zugehörigen Werte für $\cos(3x)$ in noch einer weiteren Liste (`y2)`.
* Verwenden Sie die Liste für $\sin(2x)$ (`y1`) als Liste der x-Werte und die Liste für $\cos(3x)$ (`y2`) als Liste der y-Werte des Graphen.


````python
import math
import matplotlib.pyplot as plt
 ### BEGIN SOLUTION
intervals = 1000
x_min = 0
x_max = 6.4
delta = (x_max - x_min) / intervals
x1_values = []
for interval in range(intervals+1): # Stützpunkte: Intervalle + 1
    x1_values.append(x_min + delta * interval)
y1_1_values = []
y1_2_values = []
for xi in x1_values:
    y1_1_values.append(math.sin(2 * xi))
    y1_2_values.append(math.cos(3 * xi))

plt.title("Schwingungsfigur")
plt.xlabel(r"$\sin(2x)$")
plt.ylabel(r"$\cos(3x)$")
plt.plot(y1_1_values, y1_2_values, color="red", linewidth=1)
plt.show()
 ### END SOLUTION
````
